```python
# Guessing Game

import random
jackpot = random.randint(1,100)

guess = int(input('guess a number'))
counter = 1
while(guess!=jackpot):
  if guess < jackpot:
    print('galat!! guess higher')
  else:
    print('galat!! guess lower')

  guess = int(input('guess a number'))
  counter += 1
else:
  print('correct guess')
  print('in', counter ,'attemps')
```

```python
import math
population = 10000;

for i in range(1,11):
  if population%10 == 0:
    print(population)
  else:
    print(math.floor(population))
  population = population * 10
  population = population/11
```
