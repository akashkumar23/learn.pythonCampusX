**Tuples**
A tuple in python is similar to list. The difference between the two is that we cannot change the elements of a tuple once it is assigned wherare we can change the elements of a list.

In short, a tuple is an immutable list.  A tuple cannot be changed in a way once it is created.

Characterstics
* Ordered
* Unchangable
* Allows duplicate

**Creating Tuples**


```python
#empty
t1 = ()
print(t1)
#creating   tuple with a single item
t2 = (2)
print(t2)
print(type(2))

t3 = (2,)
print(t3)
print(type(t3))
#homo
t4 = (1,2,3,4)
print(t4)
#hetero
t5 = (1,2,5.6,'hindi', (5,9,'englsi'))
print(t5)
#using type conversion
t6 = tuple('hello')
print(t6)
```

**Sets**

A set is an unordered collection of items. Every set elements is unique (no duplicates) and must be immutable (cannot) be changed.

However, a set itself is mutable. We can add or remove items from it.

Sets can also be used to perform mathematical set operation like union, intersection, symmetric difference, etc

Characterstics :-
* Unorderd
* Mutable
* No duplicates
* Can't contain mutable data types

**Creating a Set**

```python
#empty
s = {} # dictionary
print(s)
type(s)

s1 = set()
print(s1)
print(type(s1))

# 1D and 2D
s2 = {1,2,3}
print(s2)

# 2D not possible : we cannot have mutable datatype into a set

#homo and hetro
s3 = {1, 4.5, 'hello', True}
print(s3) #true is not printed , because True is 1
```

```python
#using type conversion
s4 = set([1,2,3])
print(s4)

#duplicates not allowed
s5 = {1,1,2,3,3,2,5} #errror nahi dega
print(s5)

#set cant have mutable items
s6 = {1,2,[4,5]}
print(s6)
```

```python
s1 = {1,2,3}
s2 = {2,3,1}
print(s1==s2)
```

**Comprehension**

```python
{i:i**2 for i in range(1,11)}
```

```python
dd = {'d':10, 't':50, 'o':89}
{i:j*0.1 for (i,j) in dd.items()}
```

```python
days = ['a','b','c','d','e']
temp = [10,50,89,20,36]
{i:j for (i,j) in zip(days,temp)}
```

```python
# using ip
p = {'a': 10, 'b': 50, 'c': 89, 'd': 20, 'e': 36}
{i:j for (i,j) in p.items() if j>20}
```
