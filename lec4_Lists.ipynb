**List Comprehension**
List comprehension provides a concise way of creating lists

newlist = [expression for item in iterable if condition == True]

Advantages of List Comprehension
* More time-efficient and space-efficient than loops
* Require fewer lines of code
* Transforms iterative statemtns into formulae

```python
l = []
for i in range(1,11):
  l.append(i)

print(l)
```

```python
l1 = [i for i in range(1,11)]
print(l1)
```

```python
v = [2,3,4]
s = -3
v = [s*i for i in v]
print(v)
```

```python
l = [1,2,3,4,5]
[i**2 for i in l]
```

```python
#print all the numbers divisible by 5 in the range of 1 to 50
[i for i in range(1,51) if i%5==0]
```

```python
#find the langugages which start with letter p
l = ['java', 'python', 'php', 'c']
[i for i in l if i[0]=='p']
```

```python
#nested if with list comprehension
basket = ['apple', 'guava', 'cherry', 'banana']
my_fruits = ['apple', 'kiwi', 'grapes', 'banana']

# Q : Add new list from my_fruits and items if the fruit exists in basket and also
# starts wih 'a

[i for i in my_fruits if i in basket if i[0]=='a']
```

```python
#print a (3,3) matrix using list comprehension -> nested list comprehension
[[i*j for i in range(1,4)] for j in range(1,4)]
```

```python
#cartesian products -> list comprehension on 2 lists together
l1 = [1,2,3,4]
l2 = [5,6,7,8]

l3 = [i*j for i in l1 for j in l2]

l4 = [i*j for i in l1 for j in l2 if l2.index(j)>=l1.index(i)]
print(l3)
print(l4)
```

**2 ways to traverse a list**

```python
#itemwise
l = [5,6,7,3]
for i in l:
  print(i)
```

```python
#indexwise
for i in range(0, len(l)):
  print(l[i])
```

**ZIP**
The zip() function returns a zip object, which is an iterator of tuples where the first item in each passed iterator is paired together, and then the second item in each passes iterator are paired together.

If the passes iterators have differet lengths,  the iteratord with the least items decides the length of the new iterator.

```python
l1 = [1,2,3,4]
l2 = [-1, -2 , -3, -4]
zip(l1,l2) # return object
list(zip(l1, l2))
[i+j for i,j in zip(l1,l2)]
```
